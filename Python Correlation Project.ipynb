```python
import pandas as pd 
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 
# Adjustes the configuration of the plot we will create.

# Read the Data 

df = pd.read_csv(r'D:\Users\balsin\Downloads/movies.csv')

# looking at the data 

df.head()
```

```python
# checking for missing data 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Order our Data a little bit to see

df= df.sort_values(by='gross', inplace=False, ascending=False)

#pd.set_option('max_rows', 5)
#pd.set_option('display.max_rows',10)

df.head()
```

```python
# drop duplicates

df['company'].drop_duplicates().sort_values(ascending=False)

# df.drop_duplicates() # don't do it on original dataframe
```

```python
# Budget - high correlation
# company - high correlation

# building Scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs gross Earning')

plt.xlabel('gross earning')
plt.ylabel('film budget')

plt.show()
```

```python
# regression plot using seaborn on budget and gross

sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"red"},line_kws={"color":"Blue"})
```

```python
# looking at correlation (Correlation Matrix between all numeric columns)
#pearson, kendall, spearman

df.corr(method ='pearson')
```

```python
df.corr(method ='kendall')
```

```python
df.corr(method ='spearman')
```

```python
# high correlation b/w budget and gross found 

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df.head()
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")
plt.ylabel("Movie features")

plt.show()
```

```python
corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
