```python
# Import Libraries

import pandas as pd 
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjustes the configuration of the plot we will create.

# Read the Data 

df = pd.read_csv('../input/movies/movies.csv')


# looking at the data 

df.head()
```

```python
# checking for missing data 

for col in df.columns:
    pct_mis = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_mis))
```

```python
# building Scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs gross Earning')

plt.xlabel('gross earning')
plt.ylabel('film budget')

plt.show()
```

```python
# regression plot using seaborn on budget and gross

sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
# looking at correlation (correlation matrix)

df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# high correlation b/w budget and gross found 

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix for numeric values')

plt.xlabel('features')
plt.ylabel('movie features')

plt.show()
```

```python
# now finding correlation for company  

df.head()
```

```python
# converting the company to values
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix for numeric values')

plt.xlabel('features')
plt.ylabel('movie features')

plt.show()
```

```python
corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
